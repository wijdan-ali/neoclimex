```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Import clean data file
data = pd.read_csv('data_files/clean_data.csv')
```

```python
# Roughly classifying planets based on hotness for bar graph
plt_type = ['none' for i in range(len(data))]
data['plt_type'] = plt_type
data.loc[(data["pl_eqt"] > 320), 'plt_type'] = 'hot'
data.loc[(data["pl_eqt"] >= 230) & (data["pl_eqt"] <= 320), 'plt_type'] = 'temperate'
data.loc[(data["pl_eqt"] < 230), 'plt_type'] = 'cold'

hot = data[(data.plt_type == 'hot')]
temperate = data[(data.plt_type == 'temperate')]
cold = data[(data.plt_type == 'cold')]
```

```python
# Frequency of hot, temperate, cold planets in training data | Bar Graph 
count = [len(hot), len(temperate), len(cold)]
name = ["Hot Planets", "Temperate Earths", "Cold Planets"]
colors = ["lightsalmon", "mediumaquamarine", "cornflowerblue"]

plt.figure(figsize=(6,9))
plt.bar(name, count, color = colors)
plt.yticks(np.arange(0, 11400, 300))
plt.xticks(rotation = 5)
plt.ylabel("Number Of Exoplanets")
for label in plt.gca().get_xticklabels():
    label.set_fontweight('bold')
plt.tight_layout()
plt.show()
```

```python
# Equilibrium Temperature w.r.t Average Orbital Distance | Line Graph & Scatter Plot 
data.sort_values(by='pl_orbsmax', ascending=False, inplace=True)

plt.figure(figsize=(5,3))
plt.scatter(data["pl_orbsmax"], data["pl_eqt"], s = 20, c = "tomato", alpha = 0.6)
plt.xlabel ("Orbital Distance (AU)")
plt.ylabel("Equilibrium Temperature (Teq)")
plt.grid(True)
plt.show()

plt.figure(figsize=(5,3))
plt.plot(data["pl_orbsmax"], data["pl_eqt"], alpha = 0.5)
plt.xlabel ("Orbital Distance (AU)")
plt.ylabel("Equilibrium Temperature (Teq)")
plt.grid(True)
plt.show()
```
